```python
#visualisation
import matplotlib.pyplot as plt
import numpy as np
# constantes
from prevision.pre_processing.constante import *
#data
from prevision.pre_processing.pre_process import pre_process
#models
from prevision.model.rnn_model import modele_rnn as RNN
from prevision.model.xgboost_model import general as Xgboost
```

```python
features=[x for x in ALL_FEATURES if not x.startswith("vente_") ]
X, Y = pre_process.get_data(features)
options = RNN.Options(**RNN_PARAMS_LSTM)
options.dataset_split=0.7
X_train, X_test, y_train, y_test, prevision_cage = pre_process.split(X, Y,test_size=1-options.dataset_split,random_state = 0,shuffle=False)
```

```python
#traning RNN MODEL
rnn = RNN(X, Y, options=options)
modelRnn,_ = rnn.train(plot=False)
rnn.test(modelRnn,None ,5)
res_rnn,_=rnn.compare(X,Y,modelRnn)
```

```python
#training XGBoost
modelXGBoost=Xgboost.train(X_train, X_test, y_train, y_test)
```
